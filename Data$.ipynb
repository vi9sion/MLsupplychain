```python
import pandas as pd

# Load the dataset
file_path = 'Data_Co.csv'
data = pd.read_csv(file_path, encoding='latin1')
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Checking for missing values in the dataset
missing_values = data.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

Handling Missing Values

```python
# Filling missing values with placeholders
data['Customer Lname'].fillna('Unknown', inplace=True)
data['Customer Zipcode'].fillna('Unknown', inplace=True)

# Given the large number of missing values in 'Order Zipcode' and 'Product Description',
# we can drop these columns for now as they might not be critical for our analysis.
data.drop(columns=['Order Zipcode', 'Product Description'], inplace=True)

# Verify if the missing values are handled
missing_values_post = data.isnull().sum()

# Display the result to confirm no missing values are left
missing_values_post[missing_values_post > 0]
```


# EDA

```python
import matplotlib.pyplot as plt

# Distribution of order types
plt.figure(figsize=(10, 6))
data['Type'].value_counts().plot(kind='bar')
plt.title('Distribution of Order Types')
plt.xlabel('Order Type')
plt.ylabel('Count')
plt.show()

# Sales distribution
plt.figure(figsize=(10, 6))
data['Sales'].plot(kind='hist', bins=50, logy=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency (log scale)')
plt.show()

# Distribution of delivery status
plt.figure(figsize=(10, 6))
data['Delivery Status'].value_counts().plot(kind='bar')
plt.title('Distribution of Delivery Status')
plt.xlabel('Delivery Status')
plt.ylabel('Count')
plt.show()


```

# Aggregate Sales by Category: This shows the total sales for each product category

```python
import matplotlib.pyplot as plt

# Aggregate sales by category
sales_by_category = data.groupby('Category Name')['Sales per customer'].sum().sort_values(ascending=False)

# Plotting the aggregate sales by category
plt.figure(figsize=(14, 10))
sales_by_category.plot(kind='bar', color='skyblue')
plt.title('Aggregate Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Profit of Orders: This displays the distribution of profit across different orders

```python
# Plotting the profit of orders
plt.figure(figsize=(10, 6))
plt.hist(data['Order Profit Per Order'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Order Profit Per Order')
plt.ylabel('Frequency')
plt.title('Profit of Orders')
plt.tight_layout()

# Display the plot
plt.show()
```

# Total Sales for All Markets: This visualizes the total sales in different markets

```python
# Total sales for all markets
sales_by_market = data.groupby('Market')['Sales'].sum().sort_values(ascending=False)

# Plotting the total sales for all markets
plt.figure(figsize=(10, 6))
sales_by_market.plot(kind='bar', color='orange')
plt.title('Total Sales for All Markets')
plt.xlabel('Market')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Total Sales for All Regions: This shows the total sales in different regions

```python
# Total sales for all regions
sales_by_region = data.groupby('Order Region')['Sales'].sum().sort_values(ascending=False)

# Plotting the total sales for all regions
plt.figure(figsize=(20, 14))
sales_by_region.plot(kind='bar', color='orange')
plt.title('Total Sales for All Regions')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Products with Most Loss: This identifies the top 10 products that have incurred the most loss.

```python
# Calculate the loss for each product
data['Loss'] = data['Order Item Product Price'] - data['Order Item Total']

# Aggregate loss by product
product_loss = data.groupby('Product Name')['Loss'].sum().reset_index()

# Get the top products with most loss
top_product_loss = product_loss.sort_values(by='Loss', ascending=True).head(10)

# Plotting the products with most loss
plt.figure(figsize=(12, 6))
plt.bar(top_product_loss['Product Name'], top_product_loss['Loss'], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Total Loss')
plt.title('Top 10 Products with Most Loss')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Products with most loss
loss_by_product = data[data['Order Profit Per Order'] < 0].groupby('Product Name')['Order Profit Per Order'].sum().sort_values()

# Plotting the products with most loss
plt.figure(figsize=(10, 6))
loss_by_product.head(20).plot(kind='bar', color='red')
plt.title('Top 20 Products with Most Loss')
plt.xlabel('Product Name')
plt.ylabel('Total Loss')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Top 10 Products with Late Deliveries: This identifies the top 10 products with the most late deliveries.

```python
# Filter data for late deliveries
late_deliveries = data[data['Delivery Status'] == 'Late delivery']

# Aggregate late deliveries by product
product_late_deliveries = late_deliveries.groupby('Product Name')['Order Id'].count().reset_index()
product_late_deliveries.columns = ['Product Name', 'Late Deliveries']

# Get the top 10 products with late deliveries
top_product_late_deliveries = product_late_deliveries.sort_values(by='Late Deliveries', ascending=False).head(10)

# Plotting the top 10 products with late deliveries
plt.figure(figsize=(12, 6))
plt.bar(top_product_late_deliveries['Product Name'], top_product_late_deliveries['Late Deliveries'], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Number of Late Deliveries')
plt.title('Top 10 Products with Late Deliveries')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```

# Top 10 Products with Late Deliveries: This identifies the top 10 products with the most late deliveries.

```python
# Aggregate late deliveries by region
region_late_deliveries = late_deliveries.groupby('Order Region')['Order Id'].count().reset_index()
region_late_deliveries.columns = ['Region', 'Late Deliveries']

# Get the top 10 regions with late deliveries
top_region_late_deliveries = region_late_deliveries.sort_values(by='Late Deliveries', ascending=False).head(10)

# Plotting the top 10 regions with late deliveries
plt.figure(figsize=(12, 6))
plt.bar(top_region_late_deliveries['Region'], top_region_late_deliveries['Late Deliveries'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Number of Late Deliveries')
plt.title('Top 10 Regions with Most Late Deliveries')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```

# Top 20 Products with Highest Fraud Detection

```python
# Filter data for fraudulent orders using 'Order Status'
fraud_data = data[data['Order Status'] == 'SUSPECTED_FRAUD']

# Aggregate fraudulent orders by product
product_fraud = fraud_data.groupby('Product Name')['Order Id'].count().reset_index()
product_fraud.columns = ['Product Name', 'Fraudulent Orders']

# Get the top 20 products with highest fraudulent orders
top_product_fraud = product_fraud.sort_values(by='Fraudulent Orders', ascending=False).head(20)

# Plotting the top 20 products with highest fraudulent orders
plt.figure(figsize=(12, 6))
plt.bar(top_product_fraud['Product Name'], top_product_fraud['Fraudulent Orders'], color='orange')
plt.xlabel('Product Name')
plt.ylabel('Number of Fraudulent Orders')
plt.title('Top 20 Products with Highest Fraudulent Orders')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

# Display the plot
plt.show()

# Aggregate fraudulent orders by region
region_fraud = fraud_data.groupby('Order Region')['Order Id'].count().reset_index()
region_fraud.columns = ['Region', 'Fraudulent Orders']

# Get the regions with highest fraudulent orders
top_region_fraud = region_fraud.sort_values(by='Fraudulent Orders', ascending=False).head(10)

# Plotting the regions with highest fraudulent orders
plt.figure(figsize=(12, 6))
plt.bar(top_region_fraud['Region'], top_region_fraud['Fraudulent Orders'], color='orange')
plt.xlabel('Region')
plt.ylabel('Number of Fraudulent Orders')
plt.title('Top 10 Regions with Highest Fraudulent Orders')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```
