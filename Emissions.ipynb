# Logistic Regression Model
Logistic regression model to classify sectors based on their supply chain emission factors.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Create the binary target variable
top_10_sectors['High_Emitter'] = top_10_sectors['Supply Chain Emission Factors with Margins'] > 5

# Define features (we'll use the emission factors without margins and margins as features)
X = top_10_sectors[['Supply Chain Emission Factors without Margins', 'Margins of Supply Chain Emission Factors']]
y = top_10_sectors['High_Emitter'].astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the logistic regression model
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Predict on the test set
y_pred = log_reg.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

accuracy, report
```

# Simulate emissions trends using economic growth

```python
import matplotlib.pyplot as plt

# Define different economic growth rates to test
growth_rates = [0.01, 0.03, 0.05]  # 1%, 3%, 5% annual growth

# Initialize dataframe for storing all projections
all_projections = pd.DataFrame()

# Loop through each sector and each growth rate to calculate projections
for growth_rate in growth_rates:
    for i, row in top_10_sectors.iterrows():
        sector = row['2017 NAICS Title']
        current_emissions_factor = row['Supply Chain Emission Factors with Margins']

        # Initialize lists to store yearly data
        years_list = []
        economic_activity_list = []
        emissions_list = []

        # Starting point
        initial_activity = 1  # Assume initial economic activity is 1 unit

        for year in range(years):
            current_year = 2021 + year
            current_activity = initial_activity * ((1 + growth_rate) ** year)
            current_emissions = current_activity * current_emissions_factor

            years_list.append(current_year)
            economic_activity_list.append(current_activity)
            emissions_list.append(current_emissions)

        # Create a dataframe for the sector and growth rate
        sector_df = pd.DataFrame({
            'Year': years_list,
            'Sector': sector,
            'Economic Growth Rate': growth_rate,
            'Economic Activity (unit)': economic_activity_list,
            'Projected Emissions (kg CO2e)': emissions_list
        })

        # Append to the all_projections dataframe
        all_projections = pd.concat([all_projections, sector_df], ignore_index=True)

# Plotting the trends
plt.figure(figsize=(14, 10))

for growth_rate in growth_rates:
    for sector in top_10_sectors['2017 NAICS Title']:
        subset = all_projections[(all_projections['Economic Growth Rate'] == growth_rate) & 
                                 (all_projections['Sector'] == sector)]
        plt.plot(subset['Year'], subset['Projected Emissions (kg CO2e)'], label=f"{sector} - {growth_rate*100}% growth")

plt.title('Projected Emissions Trends for Different Economic Growth Rates')
plt.xlabel('Year')
plt.ylabel('Projected Emissions (kg CO2e)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

summary_points = all_projections[(all_projections['Year'] == 2030) & 
                                 (all_projections['Sector'].isin(top_10_sectors['2017 NAICS Title']))]

summary_points
```

```python
import matplotlib.pyplot as plt

# Define different growth rates to test
growth_rates = [0.01, 0.03, 0.05]  # 1%, 3%, 5% annual growth

# Initialize a dictionary to store projections for each growth rate
projections_dict = {}

# Loop through each growth rate to calculate projections
for rate in growth_rates:
    projections = pd.DataFrame()
    for i, row in top_10_sectors.iterrows():
        sector = row['2017 NAICS Title']
        current_emissions_factor = row['Supply Chain Emission Factors with Margins']
        
        # Initialize lists to store yearly data
        years_list = []
        economic_activity_list = []
        emissions_list = []
        
        # Starting point
        initial_activity = 1  # Assume initial economic activity is 1 unit
        
        for year in range(years):
            current_year = 2021 + year
            current_activity = initial_activity * ((1 + rate) ** year)
            current_emissions = current_activity * current_emissions_factor
            
            years_list.append(current_year)
            economic_activity_list.append(current_activity)
            emissions_list.append(current_emissions)
        
        # Create a dataframe for the sector
        sector_df = pd.DataFrame({
            'Year': years_list,
            'Sector': sector,
            'Economic Activity (unit)': economic_activity_list,
            'Projected Emissions (kg CO2e)': emissions_list,
            'Growth Rate': rate
        })
        
        # Append to the projections dataframe
        projections = pd.concat([projections, sector_df], ignore_index=True)
    
    # Store the projections for the current growth rate
    projections_dict[rate] = projections

# Plot the projections for each sector and growth rate
plt.figure(figsize=(15, 10))

# Plot for each sector
for sector in top_10_sectors['2017 NAICS Title']:
    plt.subplot(2, 5, list(top_10_sectors['2017 NAICS Title']).index(sector) + 1)
    for rate in growth_rates:
        sector_data = projections_dict[rate][projections_dict[rate]['Sector'] == sector]
        plt.plot(sector_data['Year'], sector_data['Projected Emissions (kg CO2e)'], label=f'Growth Rate: {rate*100}%')
    plt.title(sector)
    plt.xlabel('Year')
    plt.ylabel('Projected Emissions (kg CO2e)')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Define policy scenarios with reduction targets
moderate_policy_reduction = 0.02  # 2% annual reduction
aggressive_policy_reduction = 0.05  # 5% annual reduction

# Initialize dictionaries to store projections under policy scenarios
moderate_policy_projections = {}
aggressive_policy_projections = {}

# Function to apply policy impacts
def apply_policy(projections, reduction_rate):
    policy_projections = pd.DataFrame()
    for sector in projections['Sector'].unique():
        sector_data = projections[projections['Sector'] == sector].copy()
        sector_data['Projected Emissions (kg CO2e)'] *= (1 - reduction_rate) ** np.arange(len(sector_data))
        policy_projections = pd.concat([policy_projections, sector_data], ignore_index=True)
    return policy_projections

# Apply policy impacts
for rate in growth_rates:
    moderate_policy_projections[rate] = apply_policy(projections_dict[rate], moderate_policy_reduction)
    aggressive_policy_projections[rate] = apply_policy(projections_dict[rate], aggressive_policy_reduction)

# Plot the projections under policy scenarios
plt.figure(figsize=(20, 15))

# Plot for each sector
for sector in top_10_sectors['2017 NAICS Title']:
    plt.subplot(2, 5, list(top_10_sectors['2017 NAICS Title']).index(sector) + 1)
    for rate in growth_rates:
        original_data = projections_dict[rate][projections_dict[rate]['Sector'] == sector]
        moderate_data = moderate_policy_projections[rate][moderate_policy_projections[rate]['Sector'] == sector]
        aggressive_data = aggressive_policy_projections[rate][aggressive_policy_projections[rate]['Sector'] == sector]
        
        plt.plot(original_data['Year'], original_data['Projected Emissions (kg CO2e)'], label=f'No Policy (Growth Rate: {rate*100}%)')
        plt.plot(moderate_data['Year'], moderate_data['Projected Emissions (kg CO2e)'], label=f'Moderate Policy (Growth Rate: {rate*100}%)', linestyle='--')
        plt.plot(aggressive_data['Year'], aggressive_data['Projected Emissions (kg CO2e)'], label=f'Aggressive Policy (Growth Rate: {rate*100}%)', linestyle='-.')
        
    plt.title(sector)
    plt.xlabel('Year')
    plt.ylabel('Projected Emissions (kg CO2e)')
    plt.legend()

plt.tight_layout()
plt.show()
```

# Analyze the impact of technological advancements on emissions

```python
# Define efficiency improvement rates
moderate_advancement = 0.02  # 2% annual improvement
high_advancement = 0.05  # 5% annual improvement

# Initialize dictionaries to store projections under technological advancement scenarios
moderate_advancement_projections = {}
high_advancement_projections = {}

# Function to apply technological advancements
def apply_technological_advancements(projections, improvement_rate):
    tech_projections = pd.DataFrame()
    for sector in projections['Sector'].unique():
        sector_data = projections[projections['Sector'] == sector].copy()
        sector_data['Projected Emissions (kg CO2e)'] *= (1 - improvement_rate) ** np.arange(len(sector_data))
        tech_projections = pd.concat([tech_projections, sector_data], ignore_index=True)
    return tech_projections

# Apply technological advancements
for rate in growth_rates:
    moderate_advancement_projections[rate] = apply_technological_advancements(projections_dict[rate], moderate_advancement)
    high_advancement_projections[rate] = apply_technological_advancements(projections_dict[rate], high_advancement)

# Plot the projections under technological advancement scenarios
plt.figure(figsize=(20, 15))

# Plot for each sector
for sector in top_10_sectors['2017 NAICS Title']:
    plt.subplot(2, 5, list(top_10_sectors['2017 NAICS Title']).index(sector) + 1)
    for rate in growth_rates:
        original_data = projections_dict[rate][projections_dict[rate]['Sector'] == sector]
        moderate_data = moderate_advancement_projections[rate][moderate_advancement_projections[rate]['Sector'] == sector]
        high_data = high_advancement_projections[rate][high_advancement_projections[rate]['Sector'] == sector]
        
        plt.plot(original_data['Year'], original_data['Projected Emissions (kg CO2e)'], label=f'No Tech Adv. (Growth Rate: {rate*100}%)')
        plt.plot(moderate_data['Year'], moderate_data['Projected Emissions (kg CO2e)'], label=f'Moderate Tech Adv. (Growth Rate: {rate*100}%)', linestyle='--')
        plt.plot(high_data['Year'], high_data['Projected Emissions (kg CO2e)'], label=f'High Tech Adv. (Growth Rate: {rate*100}%)', linestyle='-.')
        
    plt.title(sector)
    plt.xlabel('Year')
    plt.ylabel('Projected Emissions (kg CO2e)')
    plt.legend()

plt.tight_layout()
plt.show()
```
